# Create Master Datasheet for Pediatric Comfort Scores

# Import Libraries

```python
import pandas as pd
```

# Import Data
- Each file contains the mean comfort score for each stimulus for one participant

```python
# Create a list that contains all subject names (P001 to P047)
subjects = [f"P{str(i).zfill(3)}" for i in range(1, 48)]

comfort = [None] * len(subjects)

# Load comfort data for each subject
for sub in subjects:
    try:
        comfort[subjects.index(sub)] = pd.read_csv(f"Data\\PedsData\\Bracket\\{sub}\\{sub}_mean_absolute.csv")
    except FileNotFoundError:
        comfort[subjects.index(sub)] = None
        print(f"Comfort data not found for subject {sub}")
```

# Reorganize Stimulus Order to be Consistent in All Files

```python
# Define column order
stim_order = ['Contrast1Size1', 'Contrast1Size2', 'Contrast1Size3','Contrast2Size1', 'Contrast2Size2', 'Contrast2Size3',
              'Contrast3Size1', 'Contrast3Size2', 'Contrast3Size3','Contrast4Size1', 'Contrast4Size2', 'Contrast4Size3']

# Reorder columns for all DataFrames in the comfort list
for i, df in enumerate(comfort):
    if df is None:
        print(f"Skipping subject at index {i} due to missing data.")
    else:
        comfort[i] = df[stim_order]
```

# Export Comfort Scores in a Single csv

```python
combined_data = []

for i, (subject, df) in enumerate(zip(subjects, comfort)):
    row_data = {'Subject': subject}
    
    if df is not None and not df.empty:
        for col in stim_order:
            row_data[col] = df.iloc[0].get(col, None)
    else:
        # Fill with NaN for missing data
        for col in stim_order:
            row_data[col] = None
    
    combined_data.append(row_data)

# Create DataFrame
combined_df = pd.DataFrame(combined_data)

# Save to CSV
combined_df.to_csv("Data\\PedsData\\Bracket\\comfort_scores.csv", index=False)
```
