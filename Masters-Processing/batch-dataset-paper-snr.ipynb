# Filter Data
First notch filter the data to remove powerline noise (North America = 60 Hz)

Apply 7-70 Hz bandpass filter to be in line with [Benchmark SSVEP Dataset](https://doi.org/10.1109/TNSRE.2016.2627556).

```python
# Settings
hpf_fc = 7         # High-pass cut-off frequency (Hz)
lpf_fc = 70        # Low-pass cut-off frequency (Hz)
notch_freq = 60.0  # Notch filter frequency (Hz)

for i in range(len(mne_raws)):
    # Apply notch filter at 60 Hz
    n_filt_raw = mne_raws[i].copy().notch_filter(notch_freq, picks=ch_names)  

    # Apply bandpass filter
    filt_raw = n_filt_raw.copy().filter(l_freq=hpf_fc, h_freq=lpf_fc, picks=ch_names)

    # Store filtered data
    mne_raws[i] = filt_raw
```

# Epoch Data

```python
# Settings
list_of_events = []
for x in range(4):
    for y in range(3):
        list_of_events.append(f"Contrast{x+1}Size{y+1}")

# Create a dictionary of stimuli using the unique events/stimuli
dict_of_stimuli = {i: event for i, event in enumerate(list_of_events)}

# Initialize storage for batch processing
all_eeg_epochs_organized = []
all_markers_ts = []
all_markers = []
all_events_epochs = []
all_eeg_epochs = []
file_subjects = []

# Process Each File
for f, file_name in enumerate(files):
    print(f"\n[{f+1}/{len(files)}] Processing: {file_name}")
    
    # Extract subject ID
    sub = file_name.split('_')[0]
    
    if sub in subject_ids:
        try:
            # Get eeg data for this file
            filt_raw_current = mne_raws[f]
            eeg_ts_current = eeg_ts[f]
            
            # Import markers for this file 
            marker_ts, markers = import_data.read_xdf_unity_markers(f"Data\\PedsData\\EEG\\{sub}\\ses-S001\\eeg\\{file_name}")
            
            # Create epochs
            events_epochs, eeg_epochs = data_tools.create_epochs(
                eeg_data=filt_raw_current.get_data(), 
                eeg_ts=eeg_ts_current,
                markers=markers,
                markers_ts=marker_ts,
                events=list_of_events,
                epoch_end="getting score"
            )
            
            # Organize by stimulus
            eeg_epochs_organized = data_tools.epochs_stim(
                eeg_epochs=eeg_epochs,
                labels=events_epochs,
                stimuli=dict_of_stimuli
            )
            
            # Store results
            all_eeg_epochs_organized.append(eeg_epochs_organized)
            all_markers_ts.append(marker_ts)
            all_markers.append(markers)
            all_events_epochs.append(events_epochs)
            all_eeg_epochs.append(eeg_epochs)
            file_subjects.append(sub)
                        
        except Exception as e:
            print(f"ERROR: {str(e)}")
            all_eeg_epochs_organized.append(None)
            file_subjects.append(sub)

print("Total Subjects Processed:", len(file_subjects))
```

# Extract C1S1 epochs, Trim Epochs to Minimum Length, and Average Across Participants

```python
# Extract first stimulus epochs for all participants
first_stimulus_epochs_all = []

for eeg_epochs_organized in all_eeg_epochs_organized:
    if eeg_epochs_organized is not None:
        # Get all trials for Contrast1Size1
        first_stimulus_epochs = eeg_epochs_organized[0]
        if len(first_stimulus_epochs) > 0:
            first_stimulus_epochs_all.append(first_stimulus_epochs)

# Trim Epochs to minimum length
trimmed_epochs = emily_helpers.trim_epochs(first_stimulus_epochs_all)

# Extract Oz channel data from each participant's epochs. With shape [trials, channels, samples], Oz channel is at index 1 (middle dimension)
OZ_CHANNEL_INDEX = 14 

trimmed_oz_epochs = []

for participant_epochs in trimmed_epochs:
    if participant_epochs is not None:
        # Extract Oz channel: all trials, specific channel, all samples
        oz_data = participant_epochs[:, OZ_CHANNEL_INDEX, :]  # Shape: [trials, samples]
        trimmed_oz_epochs.append(oz_data)
    else:
        trimmed_oz_epochs.append(None)

# Filter out None values
valid_oz_epochs = [oz for oz in trimmed_oz_epochs if oz is not None]

# Average all Contrast1Size1 Oz data across trials for all participants
if valid_oz_epochs:
    # Concatenate all trials from all participants
    all_oz_trials = np.concatenate(valid_oz_epochs, axis=0)
    
    # Average across all trials to get the grand average
    grand_average_oz = np.mean(all_oz_trials, axis=0)
    
    print(f"\nGrand Average Oz for C1S1:")
    print(f"Shape: {grand_average_oz.shape}")  # Should be [samples]
    print(f"Length: {len(grand_average_oz)} samples")
    
    # Also calculate the standard error
    std_error = np.std(all_oz_trials, axis=0) / np.sqrt(all_oz_trials.shape[0])
    
    # For visualization or further analysis
    print(f"\nStatistics of grand average:")
    print(f"Min: {grand_average_oz.min():.4f}")
    print(f"Max: {grand_average_oz.max():.4f}")
    print(f"Mean: {grand_average_oz.mean():.4f}")
    print(f"Std: {grand_average_oz.std():.4f}")
    
else:
    print("No valid Oz epochs found.")
```
